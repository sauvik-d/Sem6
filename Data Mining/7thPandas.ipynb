```python
new_row = pd.DataFrame({'YearsExperience':[11], 'Salary': [130000], 'Name': ['BB']})

df = pd.concat([df, new_row], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(14,10))
# Scatter Plot
#plt.subplot(2, 2, 1)
plt.scatter(df['YearsExperience'], df['Salary'], color='blue')
plt.title('Scatter Plot: YearsExperience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
# bar graph using Pandas
plt.bar(df['Name'], df['Salary'], color='blue')
plt.title('Bar graph: Name vs Salary')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
df.drop(['Name',], axis=1, inplace=True)
```

```python
import numpy as np

X = np.array(df['YearsExperience'])
y = np.array(df['Salary'])
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
lr.fit(X_train, y_train)
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.plot(X_test, lr.predict((X_test)), color='r')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```
