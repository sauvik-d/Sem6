2. Write Bresenhum Line drawing Algorithm using Python.

```python
import matplotlib.pyplot as plt
```

```python
def bresenham(x0, y0, x1, y1):
    
    points = []

    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    
    sx = 1 if x0 < x1 else -1
    sy = 1 if y0 < y1 else -1
    
    err = dx - dy
    
    while True:
        points.append((x0, y0))
        
        if x0 == x1 and y0 == y1:
            break
        
        err2 = err * 2
        
        if err2 > -dy:
            err -= dy
            x0 += sx
        if err2 < dx:
            err += dx
            y0 += sy
            
    return points
```

```python
def draw_line(points):
    
    x_values, y_values = zip(*points)
    
    plt.plot(x_values, y_values, marker='o', mec='blue', mfc='blue')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Bresenham Line Drawing Algorithm')
    plt.grid(True)
    plt.show()
```

```python
x0 = int(input("Enter the X-coordinate of the first point (x1, y1): "))
y0 = int(input("Enter the Y-coordinate of the first point (x1, y1): "))

x1 = int(input("Enter the X-coordinate of the second point (x1, y1): "))
y1 = int(input("Enter the Y-coordinate of the second point (x2, y2): "))

points = bresenham(x0, y0, x1, y1)
draw_line(points)
```
