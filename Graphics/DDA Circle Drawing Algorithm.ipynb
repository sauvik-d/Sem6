```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dda_circle(x_center, y_center, radius):
    points = []

    x = 0
    y = radius
    d = 1 - radius

    while x <= y:

        points.append((x_center + x, y_center + y))  
        points.append((x_center - x, y_center + y))  
        points.append((x_center + x, y_center - y))  
        points.append((x_center - x, y_center - y))  
        points.append((x_center + y, y_center + x))  
        points.append((x_center - y, y_center + x))  
        points.append((x_center + y, y_center - x))  
        points.append((x_center - y, y_center - x))
        
        x += 1
        
        if d < 0:
            d += 2 * x + 1
        else:
            y -= 1
            d += 2 * (x - y) + 1
            
    return points
```

```python
def draw_circle(points):

    x_values, y_values = zip(*points)
    
    plt.plot(x_values, y_values, marker='o', mec='orange', mfc='orange')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('DDA Circle Drawing Algorithm')
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
x_center, y_center = map(int, input("Enter the center coordinates of the circle (x, y): ").split())
radius = int(input("Enter the radius of the circle: "))

points = dda_circle(x_center, y_center, radius)
draw_circle(points)
```
