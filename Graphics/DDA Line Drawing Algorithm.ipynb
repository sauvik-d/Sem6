1. Write DDA (Digital Differential Analyzer) Line drawing Algorithm using Python.

```python
import matplotlib.pyplot as plt
```

```python
def DDA(x1, y1, x2, y2):
    
    dx = x2 - x1
    dy = y2 - y1

    steps = abs(dx) if abs(dx) > abs(dy) else abs(dy)
    
    x_increment = dx / steps
    y_increment = dy / steps
    
    x = x1
    y = y1
    
    points = []
    
    for i in range(steps + 1):
        points.append((round(x), round(y)))
        x += x_increment
        y += y_increment
    
    return points
```

```python
def draw_line(points):
    
    x_values, y_values = zip(*points)
    
    plt.plot(x_values, y_values, marker='o', mec='red', mfc='red')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('DDA Line Drawing Algorithm')
    plt.grid(True)
    plt.show()
    
```

```python
x1 = int(input("Enter the X-coordinate of the first point (x1, y1): "))
y1 = int(input("Enter the Y-coordinate of the first point (x1, y1): "))

x2 = int(input("Enter the X-coordinate of the second point (x1, y1): "))
y2 = int(input("Enter the Y-coordinate of the second point (x2, y2): "))

points = DDA(x1, y1, x2, y2)
draw_line(points)
```
