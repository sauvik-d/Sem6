```python
import matplotlib.pyplot as plt

def bresenham_circle(x_center, y_center, radius):
    
    x = 0
    y = radius
    d = 3 - 2 * radius
    
    x_coords = []
    y_coords = []
    
    while y >= x:
        points = [(x_center + x, y_center + y), (x_center - x, y_center + y),
                  (x_center + x, y_center - y), (x_center - x, y_center - y),
                  (x_center + y, y_center + x), (x_center - y, y_center + x),
                  (x_center + y, y_center - x), (x_center - y, y_center - x)]
        
        for point in points:
            x_coords.append(point[0])
            y_coords.append(point[1])
        
        x += 1
        
        if d > 0:
            y -= 1
            d = d + 4 * (x - y) + 10
        else:
            d = d + 4 * x + 6
    
    return x_coords, y_coords

```

```python
x_center = int(input('Enter the X-coordinate of the centre: '))
y_center = int(input('Enter the Y-coordinate of the centre: '))
radius = int(input('Enter the radius of the circle: '))

x_coords, y_coords = bresenham_circle(x_center, y_center, radius)

plt.figure(figsize=(8, 8))
plt.scatter(x_coords, y_coords, s=10, color='blue')
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Bresenham's Circle Drawing Algorithm")
plt.show()
```
