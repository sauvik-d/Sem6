```python
import matplotlib.pyplot as plt
```

```python
def DDA(x1, y1, x2, y2):

    dx = x2 - x1
    dy = y2 - y1

    steps = int(max(abs(dx), abs(dy)))

    x_increment = dx / steps
    y_increment = dy / steps

    x = x1
    y = y1

    points = []
    
    for i in range(steps + 1):
        points.append((round(x), round(y)))
        x += x_increment
        y += y_increment
    
    return points
```

```python
def draw_line(points):
    # Unzip the points into x and y coordinates for plotting
    x_values, y_values = zip(*points)
    
    # Plotting the line using matplotlib
    plt.plot(x_values, y_values, marker='o', mec='red', mfc='red')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('DDA Line Drawing Algorithm')
    plt.grid(True)
    plt.show()
```

```python
# Example usage
x1 = int(input('Enter the X-coordinate of the starting point'))
y1 = int(input('Enter the Y-coordinate of the starting point'))
x2 = int(input('Enter the X-coordinate of the end point'))
y2 = int(input('Enter the Y-coordinate of the end point'))

points = DDA(x1, y1, x2, y2)
print("Points between the two points:", points)
draw_line(points)
```
