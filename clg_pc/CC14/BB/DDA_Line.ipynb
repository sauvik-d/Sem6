```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def DDA_Algo(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    step = 0
    if dx > dy:
        step = dx
    else:
        step = dy
    x_increment = dx / step
    y_increment = dy / step
    x = []
    y = []
    while(x2 > x1 and y2 > y1):
        x1 = x1 + x_increment
        y1 = y1 + y_increment
        x.append(round(x1))
        y.append(round(y1))

    return x, y
    
```

```python
x1 = int(input('Enter the x-coordinate of the starting point'))
y1 = int(input('Enter the y-coordinate of the starting point'))
x2 = int(input('Enter the x-coordinate of the end point'))
y2 = int(input('Enter the y-coordinate of the end point'))

x , y = DDA_Algo(x1, y1, x2, y2)
```

```python
plt.plot(x, y, color='blue', marker='o')
plt.title("DDA Line Drawing Algorithm")
plt.grid(True)
plt.show()
    
```
