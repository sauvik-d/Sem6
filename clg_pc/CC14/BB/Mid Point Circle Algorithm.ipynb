```python
import matplotlib.pyplot as plt
import numpy as np

def midpoint_circle(x_center, y_center, radius):

    x = 0
    y = radius
    d = 1 - radius
    
    x_coords = []
    y_coords = []
    
    while x <= y:
        x_coords.extend([x_center + x, x_center + y, x_center - y, x_center - x,
                         x_center - x, x_center - y, x_center + y, x_center + x])
        y_coords.extend([y_center + y, y_center + x, y_center + x, y_center + y,
                         y_center - y, y_center - x, y_center - x, y_center - y])
        
        x += 1
        
        if d < 0:
            d += 2 * x + 1
        else:
            d += 2 * (x - y) + 1
            y -= 1
    
    return x_coords, y_coords
```

```python
x_center = int(input('Enter the X-coordinate of the centre: '))
y_center = int(input('Enter the Y-coordinate of the centre: '))
radius = int(input('Enter the radius of the circle: '))

x_coords, y_coords = midpoint_circle(x_center, y_center, radius)

plt.figure(figsize=(8, 8))
plt.scatter(x_coords, y_coords, s=10, color='blue')
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Midpoint Circle Drawing Algorithm")
plt.show()
```
