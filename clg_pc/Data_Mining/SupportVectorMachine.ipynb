```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
df.describe()
df.info()
```

```python
df.describe()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Sepal Width and Length Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+' , label = iris.target_names[0])
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.',  label = iris.target_names[1])
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'],color="red",marker='*',  label = iris.target_names[2])
plt.legend()
plt.show()
```

```python
plt.title('Petal Width and Length Plot')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'],color="red",marker='*')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target'], axis='columns')
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC(gamma=0.5) # here the default kernel used is Gaussain Kernel
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

Support Vector Machine using Linear Kernels

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
