```python
import numpy as np
from collections import Counter
import pandas as pd
import math
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
def distance(p1, p2):
    return math.dist(p1, p2)
```

```python
def predict_KNN(X_train, y_train, predict_point, k):
    distances = []
    for i in range(len(X_train)):
        distances.append(distance(X_train[i], predict_point))
    arg_distance = np.argsort(distances)
    select_X = arg_distance[:k]
    for i in range(len(select_X)):
        y_select = y_train[select_X]
    
    return Counter(y_select).most_common(1)[0][0], (y_select)
```

```python
point = (5, 3, 3, 2)
predict_KNN(df.values[:, :-1], df.values[:, -1], point, 5)
```
